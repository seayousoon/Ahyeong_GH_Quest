```python
tags_en = nltk.pos_tag(words)
```

```python
tags_en[:20]
```

```python
# 명사 분류
Ifnouns = ['NN', 'NNS', 'NNP', 'NNPS', 'PRP', 'PRP$']
nouns = []
for a in tags_en:
    for noun in Ifnouns:
        if noun in a:
            nouns.append(a[0])
nouns[:10]
```

```python
# 동사 분류
ifverb = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
verbs = []
for a in tags_en:
    for verb in ifverb:
        if verb in a:
            verbs.append(a[0])
verbs[:10]
```

```python
# 형용사 분류
ifadjective = ['JJ', 'JJR', 'JJS']
adjectives = []
for a in tags_en:
    for adjective in ifadjective:
        if adjective in a:
            adjectives.append(a[0])
adjectives[:10]
```

```python
import matplotlib.pyplot as plt
```

```python
# top20 단어 그래프 그리기

nouns_20 =nltk.Text(nouns)
nouns_20.plot(20)
plt.show()
```

```python
verbs_20 =nltk.Text(verbs)
verbs_20.plot(20)
plt.show()
```

```python
adjectives_20 = nltk.Text(adjectives)
adjectives_20.plot(20)
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel-reviews.csv')['Description']
```

```python
df = df.str.replace('[^a-zA-Z]',' ').str.lower()
df.head(10)
```

```python
df = df.str.split()
```

```python
import gensim
import multiprocessing
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(df,size=100, window=3,workers=4, min_count=5, sg=1)
```

```python
model.init_sims(replace=True)
```

```python
# 두 단어 간 유사성 
print(model.wv.similarity('staff', 'friendly'))
print(model.wv.similarity('unhappy', 'rude'))
print(model.wv.similarity('breakfast', 'morning'))
print(model.wv.similarity('breakfast', 'included'))
print(model.wv.similarity('location', 'close'))
```

```python
from sklearn.manifold import TSNE
import matplotlib as mpl
import matplotlib.pyplot as plt
import gensim 
import gensim.models as g
```

```python
mpl.rcParams['axes.unicode_minus'] = False

vocab = list(model.wv.vocab)
X = model[vocab]

print(len(X))
print(X[0][:10])
tsne = TSNE(n_components=2)

X_tsne = tsne.fit_transform(X[:150,:])
```

```python
df = pd.DataFrame(X_tsne, index=vocab[:150], columns=['x', 'y'])
df.shape
```

```python
df.head(10)
```

```python
# 단어 유사성분포 시각화

fig = plt.figure()
fig.set_size_inches(40, 20)
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df['x'], df['y'])

for word, pos in df.iterrows():
    ax.annotate(word, pos, fontsize=30)
plt.show()
```
